```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Read in data from two different outputs:

```python
df = pd.read_csv('gridtest_output.txt')
df_nona = df.dropna()

test_df = pd.read_csv('old_gridtest_output.txt')
```

#### Test that both runs yield the same general results: (should equal zero)

```python
df['worked'].size - np.count_nonzero(test_df['worked'] == df['worked'])
```

#### See how many nan rows we got:

```python
df.size-df_nona.size
```

### Correlation plot:

```python
corr = df_nona.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(df_nona.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(df_nona.columns)
ax.set_yticklabels(df_nona.columns)
ax.set_title('Correlation Plot')
```

It appears inclination has the strongest correlation w/ distance... I have little intution as to what to make of this though

### Histogram:

```python
plt.hist(df_nona['distance'], bins = 20, range = [0,3.5])
plt.yscale('log')

plt.title('Histogram of distance values (log y scale)');
```

```python
plt.scatter(df_nona['distance'], df_nona['inc'])
plt.xlabel('distance at start (AU)')
plt.ylabel('initial system inclination (rad/pi)')
```
